```python
#Table 2 Device_Access_Connectivity Table
import pandas as pd
device_access_connectivity_data = pd.read_csv('/Users/bhumikasaxena/Downloads/Device_Access_Connectivity.csv')
device_access_connectivity_data

```

```python
# Remove duplicates and remove Null Values

device_access_connectivity_data.drop_duplicates()
dev_acc_conn_data = device_access_connectivity_data.dropna()
dev_acc_conn_data
```

```python
# Table 3 Health_literacy_communication

import pandas as pd
health_literacy_communication_data = pd.read_csv('/Users/bhumikasaxena/Downloads/Health_Literacy_Communication.csv')
health_literacy_communication_data
```

```python
# Remove duplicates and remove Null Values

health_literacy_communication_data.drop_duplicates()
hea_lit_comm_data = health_literacy_communication_data.dropna()
hea_lit_comm_data
```

```python
# Table 4 Lifestyle_Access
import pandas as pd
lifestyle_access_data = pd.read_csv('/Users/bhumikasaxena/Downloads/Lifestyle_Access.csv')
lifestyle_access_data
```

### Step 3

#### Merge

```python
# Functionality 1
# MERGE

final_merge = patient_data.merge(emp_ins_data, how ='left',on='Patient_ID') \
                          .merge(dev_acc_conn_data, how = 'left', on='Patient_ID') \
                          .merge(hea_lit_comm_data, how = 'left', on='Patient_ID') \
                          .merge(lif_acc_data, how = 'left',on='Patient_ID')
final_merge
```

#### Append the Recovery Column

```python
# Functionality 2
# APPEND

def label_recovery(row):
    if row <= 20:
        return 'Fast'
    else:
        if row <= 50:
            return 'Average'
        else:
            return 'Slow'

final_merge['Recovery_Speed'] = final_merge['Recovery_Time_Days'].apply(label_recovery)
final_merge
```

```python
# Functionality 3
# Dummy Variables

# Create dummy variables
# Two separate rows with true and false with visit type
final_merge_with_dummies = pd.get_dummies(final_merge, columns=['Visit_Type'])

final_merge_with_dummies

```

```python
# Creating dummy variablewith a single row explicitly with 0 and 1 (binary variable)


final_merge['Is_Internet_Access'] = final_merge['Internet_Access'].apply(lambda x: 1 if x == 'Yes' else 0)
final_merge['Is_telehealth'] = final_merge['Visit_Type'].apply(lambda x:1 if x== 'Telehealth' else 0)
final_merge
```

```python
# Functionality 4
# Binning / Bucketing

final_merge['Age_Group'] = pd.cut(
    final_merge['Age'],
    bins=[0, 30, 50, 100],
    labels=['Young', 'Middle', 'Senior']
)

final_merge
```

```python
# Functionality 5
# Using groupby() + mean():

# Compare avg Recovery Time between Telehealth & In-Person visits

final_merge.groupby('Visit_Type')['Recovery_Time_Days'].mean()

```

```python
avg_recovery = final_merge.groupby('Visit_Type')['Recovery_Time_Days'].mean()

import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.bar(avg_recovery.index, avg_recovery.values, color='pink',edgecolor='black')
plt.title("Avg Recovery Time by Visit Type")
plt.ylabel("Recovery Time (Days)")
plt.xlabel("Visit Type")

plt.show()
```

```python
# Functionality 6
# Using value_counts():

# Count patients by Recovery Speed (Fast, Average, Slow)


final_merge['Recovery_Speed'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Count patients by Recovery Speed
recovery_counts = final_merge['Recovery_Speed'].value_counts()
colors = ["pink", "lightgreen", "steelblue"]

# Plotting a pie chart
plt.figure(figsize=(6, 6))
plt.pie(recovery_counts, labels=recovery_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)


plt.title('Distribution of Patients by Recovery Speed')

plt.tight_layout()
plt.show()
```

```python
# Functionality 7
# Get Max/Min by Group
# Max and Min Recovery Time by Visit Type

final_merge.groupby('Visit_Type')['Recovery_Time_Days'].agg(['min', 'max'])
```

```python
# Functionality 8
# Using if/else condition to generate new column
# Created a new column name Patient_Experience

def label_patient_experience(row):
    if row['Satisfaction_Level'] >= 4 and row['Recovery_Speed'] == 'Fast':
        return 'Great'
    else:
        if row['Satisfaction_Level'] <= 2 and row['Recovery_Speed'] == 'Slow':
            return 'Poor'
        else:
            return 'Average'

final_merge['Patient_Experience'] = final_merge.apply(label_patient_experience, axis=1)
final_merge
```

```python
counts = final_merge['Patient_Experience'].value_counts()
counts.plot.pie(autopct='%1.1f%%', startangle=90, figsize=(6, 6), colors=['lightgreen', 'gold', 'lightcoral'])

plt.title('Patient Experience Distribution')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Created a new column name lifestyle_score

def lifestyle_score(row):
    if row['Exercise_Frequency'] == 'High' and row['Diet_Quality'] == 'Good':
        return 'Healthy'
    else:
        if row['Exercise_Frequency'] == 'Low' or row['Diet_Quality'] == 'Poor':
            return 'Unhealthy'
        else:
            return 'Moderate'

final_merge['Lifestyle_Score'] = final_merge.apply(lifestyle_score, axis=1)
final_merge
```

## Inferential statistics

```python
# 1. Correlation

final_merge[['Recovery_Time_Days', 'Satisfaction_Level', 'Treatment_Cost']].corr()
```

```python
#  2. Linear Regression
# Predicting treatment cost from recovery time

import statsmodels.api as sm

X = final_merge[['Recovery_Time_Days']]
X = sm.add_constant(X)
y = final_merge['Treatment_Cost']

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
plt.figure(figsize=(8,5))
plt.scatter(final_merge['Recovery_Time_Days'], final_merge['Treatment_Cost'], label='Data Points')
plt.plot(final_merge['Recovery_Time_Days'], model.predict(X), color='red', label='Regression Line')
plt.xlabel('Recovery Time (Days)')
plt.ylabel('Treatment Cost')
plt.title('Linear Regression: Treatment Cost vs. Recovery Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# T-test

from scipy.stats import ttest_ind

tele = final_merge[final_merge['Visit_Type'] == 'Telehealth']['Recovery_Time_Days']
in_person = final_merge[final_merge['Visit_Type'] == 'In-Person']['Recovery_Time_Days']

t_stat, p_val = ttest_ind(tele, in_person)
print(f"T-statistic: {t_stat}, P-value: {p_val}")
```
